```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a=pd.read_csv('booking.csv')
a
b=pd.read_csv('users.csv')
b
c=pd.read_csv('clinics.csv')
c
d=pd.read_csv('procedures.csv')
d
e=pd.read_csv('inquries.csv')
d.info()
c
```

```python
f=pd.merge(d,e,on='procedure_id',how='inner')
f
```

```python
f["inquirescost"]=f["procedure_name"].replace({'Breast Implants':100,'Teeth Whitening':50,'Hip Replacement':250,'Veneers':45,'Cancer Treatment':300,'Laser Eye Surgery':75})
f


```

```python
g=pd.merge(f,a,on='inquiry_id',how='inner')
g
g['procedurecost']=g['procedure_name'].replace({'Hip Replacement':10000,'Veneers':300,'Teeth Whitening':250,'Breast Implants':5000,'Cancer Treatment':9000,'Laser Eye Surgery':1000})
g
```

```python
h=pd.merge(f,b,on='user_id',how='inner')
h
i=pd.merge(g,b,on='user_id',how='inner')
i



           

```

```python
j=pd.merge(h,c,on='clinic_id',how='inner')
j
k=pd.merge(i,c,on='clinic_id',how='inner')
k

```

```python
l=j.groupby('created_at_inquires')['inquiry_id'].count().sort_values(ascending=False).head(10)
l
m=j.groupby('procedure_name')['inquiry_id'].count().sort_values(ascending=False)
m
n=j.groupby('clinic_id')['inquiry_id'].count().sort_values(ascending=False)
n
#j['clinic_location'].value_counts()
l
#plt.figure(figsize=(20,5))
#plt.plot(l.index,l.values)
#plt.setp(l.index, rotation=90)

#plt.show()
m
plt.figure(figsize=(20,8))
plt.bar(m.index,m.values,label='procedure vs inquiries',color='red')
plt.legend(loc='upper right')

plt.ylabel('number of inquiries')
plt.xlabel('procedure name')
plt.show()
m
plt.figure(figsize=(10,5))
plt.plot(m.index,m.values)
plt.xlabel('procedure name')
plt.ylabel('no. of inq')
plt.show()
```
