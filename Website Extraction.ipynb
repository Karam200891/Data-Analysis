```python
Name=[]
Price=[]
Wprice=[]
Sprice=[]
for j in link[1:]:
    print(j)
    
    pg=requests.get(j)
    pg.text
    page_html=BeautifulSoup(pg.text,'html.parser')
    main=page_html.find_all('div',class_='finner')
    
    for container in main:
        n=container.find('div',class_='name').text
        p=container.find('div',class_='strike-price')
        wp=container.find('div',class_='price')
        s=container.find('div',class_='savings')
        Name.append(n)
        Wprice.append(wp)
        Price.append(p)
        Sprice.append(s)
        
     
    
    #for k in sp.find_all('div',class_='name'):
        #print(k.text)
        
        
    #for l in sp.find_all('div',class_='strike-price'):
        #print(l.text)
        
    #for m in sp.find_all('div',class_='price'):
        #print(m.text)
       
    #for n in sp.find_all('div',class_='savings'):
        #print(n.text)
       
    
    
    
```

```python
import pandas as pd
df=pd.DataFrame({'PPName':Name,'PPPrice':Price,'PPWPrice':Wprice,'PPSprice':Sprice})
df=df.dropna()

df
```

```python
price=[]
wprice=[]
sprice=[]


for i in df["PPPrice"]:
    print(i.text)
    price.append(i.text)
for j in df["PPWPrice"]:
    print(j.text)
    wprice.append(j.text)
for l in df["PPSprice"]:
    print(l.text)
    sprice.append(l.text)
```

```python
df["Price"]=price
df
df["Wholesale"]=wprice
df
df["sellingprice"]=sprice
df.info()
```

```python
df["sp"]=df["sellingprice"].str.split(":").str.get(1)
df["spp"]=df['sp'].str.split(')').str.get(0)
df
```
